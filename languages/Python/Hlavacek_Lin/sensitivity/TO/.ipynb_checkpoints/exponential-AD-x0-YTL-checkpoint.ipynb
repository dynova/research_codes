```python
import numpy as np
import jax.numpy as jnp
import diffrax
import jax
from jax import grad, jacrev
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from matplotlib import cm
from time import time
jax.config.update("jax_enable_x64", True)
plt.rcParams.update({'font.size':10})
import gc
```

```python
r = 1.
x0 = 1./19617869
tSpan = jnp.linspace(0,1.0/r*np.log(1./x0),50001)

tol=1e-8
gc.collect()

@jax.jit
def RHS(t,x,r):  
    return r*x
    #return r # debugging example: linear dynamics 

@jax.jit
def xt(r):
    terms = diffrax.ODETerm(RHS)
    solver = diffrax.Dopri8()
    t0 = tSpan[0]
    t1 = tSpan[-1]
    dt0 = None
    IC = x0
    saveat = diffrax.SaveAt(ts=tSpan)
    stepsize_controller = diffrax.PIDController(rtol=tol, atol=tol)    
    sol = diffrax.diffeqsolve(terms,solver,t0,t1,dt0,IC,args=(r),saveat=saveat,stepsize_controller=stepsize_controller,max_steps=int(1e12))
    return sol.ys    
```

```python
sol = xt(1.0)
```

```python
plt.plot(tSpan, sol, zorder = 333)
plt.plot(tSpan, x0*jnp.exp(tSpan), zorder = 111, lw=4)
```

```python
plt.plot(tSpan, sol-x0*jnp.exp(tSpan), zorder = 111, lw=4)
```

```python
sensAD = jacrev(xt)(r)
```

```python
plt.plot(tSpan, jnp.log(sensAD), label='diffrax AD',zorder=333)
plt.plot(tSpan, jnp.log(x0)+r*tSpan,lw=4, zorder=111)
plt.xlabel('Time')
plt.ylabel('$\\log $ Sensitivity')
```

```python
plt.plot(tSpan, sensAD, label='$\\partial x(t)/\\partial r$ by JAX/DIFFRAX AD',zorder=333)
plt.plot(tSpan, r*x0*jnp.exp(r*tSpan), label='analytical $\\partial x(t)/\\partial r = r x_0 e^{t}$')
#plt.plot(tSpan, r*tSpan, label='analytical $\\partial x(t)/\\partial r = t$',lw=4, zorder=111) # debugging example: linear dynamics 
plt.xlabel('time')
plt.ylabel('Sensitivity')
plt.legend(loc=2)
```
