```python
from sympy import *
TR, TA, E1, E2, E3, IS, IL, V, VP, B1, B2, B3, B4, C3, C4, S, A, R = symbols('TR TA E1 E2 E3 IS IL V VP B1 B2 B3 B4 C3 C4 S A R')
lamb, mur, alpha, betar, gamma, mua, betaa, ke, epsilon, delta, f, ps, pl, c, kappai, ft, kappae, n, kf, kr, kpx, kmx, kpc, kmc, ka, kd, rt  = symbols('lamb mur alpha betar gamma mua betaa ke epsilon delta f ps pl c kappai ft kappae n kf kr kpx kmx kpc kmc ka kd rt')
flow = Matrix([lamb-mur*TR-alpha*V*TR-betar*V*TR,alpha*V*TR+gamma*V*TA-mua*TA-betaa*V*TA,betar*V*TR+betaa*V*TA-ke*E1,ke*E1-ke*E2,ke*E2-ke*E3,(1-epsilon)*ke*E3-delta*IS,epsilon*ke*E3-f*delta*IL,ps*IS+pl*IL-c*V-kappai*V+ft*kappae*VP-n*kf*R*V+kr*B1,kappai*V-kappae*VP,n*kf*R*V-kr*B1-((n-1)*kpx*R*B1-2*kmx*B2),(n-1)*kpx*R*B1-2*kmx*B2-((n-2)*kpx*R*B2-3*kmx*B3),(n-2)*kpx*R*B2-3*kmx*B3-((n-3)*kpx*R*B3-4*kmx*B4)-(kpc*B3-kmc*C3),(kpx*R*B3-4*kmx*B4)-(kpc*B4-kpc*C4),kpc*B3-kmc*C3,kpc*B4-kmc*C4,-ka*A*R+kd*S,0,0])
flow[-1] = -(flow[-3]+flow[9]+2*flow[10]+3*(flow[11]+flow[13])+4*(flow[12]+flow[14]))
states_all = Matrix([TR, TA, E1, E2, E3, IS, IL, V, VP, B1, B2, B3, B4, C3, C4, S, A, R])
states_infected = Matrix([E1, E2, E3, IS, IL, V, VP, B1, B2, B3, B4, C3, C4])
temp = flow.jacobian(states_all)[2:-3,2:-3].subs([(TR,lamb/mur),(TA,0),(R,rt)])
FF = zeros(len(states_infected),len(states_infected))
FF[0,5] = temp[0,5]
VV = FF-temp
R0 = list((FF*(VV**-1)).eigenvals().keys())[-1]
```
